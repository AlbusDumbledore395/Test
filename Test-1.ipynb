```python
import pandas as pd

data = {
    'Ename': ['Roshan', 'Amar', 'Ashwini', 'Lohith', 'Mohan', 'Pramod'],
    'Type': ['regular', 'adhoc', 'regular', 'adhoc', 'contract', 'regular'],
    'Dname': ['cs', 'cs', 'ec', 'ec', 'cs', 'ec'],
    'exp': [10, 20, 5, 14, 9, 8],  # Removed one value to match the number of employees
    'salary': [5000, 15000, 3000, 15000, 10000, 40000]
}

df = pd.DataFrame(data)
df
```

```python
avg_salary = df.pivot_table(values='salary', index='Ename', columns='Dname', aggfunc='mean')
avg_salary
```

```python
salary_stats=df.pivot_table(values='salary',index='Type',aggfunc={'salary':['sum','mean','count']})
salary_stats
```

# describe the different types of machine learning algorithms with examples.
Supervised Learning:

Description: Algorithms learn from labeled data (input-output pairs) to make predictions or classify new data.
Examples:
Linear Regression (predicting numeric values).
Logistic Regression (binary classification).
Decision Trees (classification and regression).
Unsupervised Learning:

Description: Algorithms find patterns and structures in unlabeled data.
Examples:
K-Means Clustering (grouping similar data points).
Principal Component Analysis (dimensionality reduction).
Gaussian Mixture Models (probability-based clustering).
Semi-Supervised Learning:

Description: A combination of supervised and unsupervised learning, where models learn from a mix of labeled and unlabeled data.
Examples:
Label Propagation (predict labels in partially labeled data).
Self-training (use labeled data to label unlabeled data).
Reinforcement Learning:

Description: Algorithms learn by interacting with an environment and receiving rewards or penalties for actions.
Examples:
Deep Q-Networks (DQN) for game playing (e.g., AlphaGo).
Proximal Policy Optimization (PPO) for robotics control

# Goals of AI.
Automate Tasks: AI aims to automate routine and labor-intensive tasks, allowing machines to perform them with speed and accuracy, reducing human effort.

Solve Complex Problems: AI strives to solve complex problems that may be beyond the capabilities of human beings, such as analyzing vast datasets or optimizing complex systems.

Mimic Human Intelligence: AI seeks to replicate and simulate human intelligence, enabling machines to understand, reason, learn, and make decisions in a human-like manner.

Enhance Decision-Making: AI intends to provide valuable insights and recommendations to assist in decision-making processes, both in business and various domains.

Improve Efficiency and Productivity: AI technologies aim to enhance efficiency, productivity, and resource utilization in various industries, leading to cost savings and improved outcomes.

Personalize Experiences: AI endeavors to personalize user experiences, from content recommendations to tailored healthcare treatment plans, based on individual preferences and needs.

Advance Research and Innovation: AI contributes to scientific discovery and innovation by accelerating research, pattern recognition, and data analysis.

Enhance Safety and Security: AI is used to develop systems for monitoring, analyzing, and ensuring the safety and security of various environments, such as autonomous vehicles and cybersecurity.

Enable Natural Language Understanding: AI focuses on enabling machines to understand and interact with humans in natural language, improving communication and accessibility.

Drive Economic Growth: AI plays a crucial role in driving economic growth by fostering innovation, creating new industries, and improving existing ones
